```python
import pandas as pd
import glob
```

```python
path = r'/Users/mariaquinton/Desktop/citibike_datasets' # use your path
all_files = glob.glob(path + "/*.csv")
```

```python
bike_data = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    bike_data.append(df)

frame = pd.concat(bike_data, axis=0, ignore_index=True)
```

```python
frame.head()
```

```python
frame.to_csv('/Users/mariaquinton/Desktop/citi_bike.csv')
```
